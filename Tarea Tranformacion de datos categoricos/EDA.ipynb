```python
levels1 = adidas.groupby(["category"], as_index = False).agg({"sku" : ["count"]})
```

```python
levels1.columns = levels1.columns.droplevel(1)
```

```python
levels1
```

```python
levels1['Percentage'] = levels1['sku']/ adidas.shape[0]
```

```python
levels1
```

```python
levels1['category_t'] = 'Others'
```

```python
levels1
```

```python
levels1['category_t'] = levels1.apply(lambda r: r['category'] if r ['Percentage'] > 0.39 else r['category_t'], axis = 1)
```

```python
levels1
```

```python
levels1 = levels1.loc[ : , ["category","category_t"]]
```

```python
levels1
```

```python
adidas = adidas.merge(levels1, on = ["category"], how = "inner" )
```

```python
adidas.groupby(["category_t"], as_index = False).agg({"sku" : ["count"]})
```
